```
def connect_nodes(string):
  jump = num_children-1
  positions = [m.start() for m in re.finditer('/', string)]
  #for i, pos in enumerate(positions):
    #print (i,pos)
  start_indexes =[i for i in range(len(positions)) if i % (num_children) == 0]
  #print(positions)
  #print(start_indexes)
  if len(start_indexes) > 0 and len(positions)>1:
    mytuples = [(positions[i],positions[i+jump]) for i in (start_indexes)]
    string_list = list(string)
    for (start,stop) in mytuples:
      #print (start,stop)
      for i in range (start+1,stop):
        string_list[i] = "-"
    newstring = "".join(string_list)
  else:
    newstring = string
  return (newstring)
```

```
def count_marks(marks):
  """
  zero_marks_list,ones_marks_list = count_marks(marks)
  utility function for tag_node_OPT
  """
  x = np.array(marks)
  zero_marks_list = list(np.where(x == 0)[0])
  ones_marks_list = list(np.where(x == 1)[0])
  return zero_marks_list,ones_marks_list
```

```
def count_assign_parent(ones_marks_list,zero_marks_list,d):
  """
  condition, parent= count_assign_parent(ones_marks_list,zero_marks_list,d)
  utility function for tag_node_OPT
  """ 
  
  # if not all the children are marked then:
  if len(ones_marks_list)+len(zero_marks_list) < num_children:
    if iseven(d):
      condition = (len(ones_marks_list) > 0)
      parent = 1
    else: 
      condition = (len(zero_marks_list) > 0)
      parent = 0
    # condition out of here can be False
  
  # if all the children have been marked
  else:
    if iseven(d):
      condition = (len(ones_marks_list) > 0)
      if condition: 
        parent = 1
      else: # if there isnt any 1
        parent = 0
    else: 
      condition = (len(zero_marks_list) > 0)
      if condition: 
        parent = 0
      else: # if there isnt any 0
        parent = 1
    # we reassign the condition True, because all children have been labelled
    condition = True


  return condition, parent
```
